```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python

apples_yeild = [0.85,0.91,0.72,0.56]
years = [2011,2012,2013,2014]
plt.xlabel('Years')
plt.ylabel('Yield (tons per hectare)')
plt.plot(years,apples_yeild)
```

```python
months = range(1,13)
mangos = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931, 0.934, 0.936, 0.937, 0.9375, 0.9372,0.78]
bananas = [0.962, 0.941, 0.930, 0.923, 0.918, 0.908, 0.907, 0.904, 0.901, 0.898, 0.9, 0.67]

plt.plot(months, mangos, marker ='o')
plt.plot(months, bananas, marker ='x')

plt.title("Crop Yields in Kanto")
plt.legend(['mangos', 'bananas'])

plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')
```

```python
years = list(range(2012, 2024))
plt.plot(years, mangos, marker='s', c='b', ls='-', lw=2, ms=8, mew=2, mec='navy')
plt.plot(years, bananas, marker='o', c='r', ls='--', lw=3, ms=10, alpha=.5)


plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')
plt.title("Crop Yields in Kanto")
plt.legend(['mangos', 'bananas'])
```

```python
plt.plot(years, mangos, 's-b')
plt.plot(years, bananas, 'o--r')
plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')
plt.title("Crop Yields in Kanto")
plt.legend(['Apples', 'Oranges'])
```

```python
plt.plot(years, bananas, 'or')
plt.title("Yield of Oranges (tons per hectare)");
```

```python
plt.figure(figsize=(12, 6))
plt.plot(years, mangos, 'or')
plt.title("Yield of Oranges (tons per hectare)")
```

```python
sns.set_style("darkgrid")

plt.plot(years, mangos, 's-b')
plt.plot(years, bananas, 'o--r')
plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')
plt.title("Crop Yields in Kanto")
plt.legend(['Apples', 'Oranges'])
```
